```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Bank_Satisfaction.csv')
```

```python
print(df.head())
```

```python
## Shape of the dataframe
```

```python
print(df.shape)
```

```python
## EDA OR DATA ANALYSIS 
```

```python
features_nan=[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
features_nan
```

```python
## lets see the numerical variables
```

```python
numerical_features=[features for features in df.columns if df[features].dtypes!='O']
```

```python
numerical_features
```

```python
print("No of numerical_features :",len(numerical_features))
```

## HERE ALL THE FEATURES ARE NUMERIC AND THERE IS NO CATEGORICAL FEATURES.

```python
## discrete features 
```

```python
discrete_features=[features for features in numerical_features if (len(df[features].unique())<25)]
```

```python
discrete_features
```

```python
print(f"The no of discrete features :{len(discrete_features)}")
```

## lets see the distribution of discrete features 

```python
for features in discrete_features:
    df[features].hist()
    plt.xlabel(features)
    plt.ylabel('Count')
    plt.title(features)
    plt.show()
```
