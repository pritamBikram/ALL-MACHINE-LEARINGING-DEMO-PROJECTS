```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Health_train.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

## EDA OR DATA ANALYSIS

```python
features_nan=[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
features_nan
```

## THERE IS NO NULL VALUE PRESENT

```python
## numerical features
```

```python
numerical_features=[features for features in df.columns if df[features].dtypes!='O']
```

```python
numerical_features
```

## discrete_features

```python
discrete_features=[features for features in numerical_features if(len(df[features].unique())<25)]
```

```python
discrete_features
```

```python
## lets see the distribution
```

```python
for features in discrete_features:
    df.groupby(features)['Response'].mean().plot.bar()
    plt.xlabel(features)
    plt.ylabel('RESPONSE')
    plt.title(features)
    plt.show()
```

```python
## lets see the distribution of the discrete_features
```

```python
for features in discrete_features:
    df[features].hist()
    plt.xlabel(features)
    plt.ylabel('COUNT')
    plt.title(features)
    plt.show()
```

```python
continuous_features=[features for features in numerical_features if features not in discrete_features]
```

```python
continuous_features
```

```python
# lets see the distribution of the features
```

```python
for features in continuous_features:
    df[features].hist()
    plt.xlabel(features)
    plt.ylabel('COUNT')
    plt.show()
    
```

```python
## lest see the droplets 
```

```python
for features in numerical_features:
    df.boxplot(column=features)
    plt.ylabel(features)
    plt.title(features)
    plt.show()
```
