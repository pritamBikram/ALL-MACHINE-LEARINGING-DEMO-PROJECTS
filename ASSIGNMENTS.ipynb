```python
import matplotlib.pyplot as plt
import math
```

```python
import random
```

```python
l=[]
for i in range(20):
    x=random.randint(0,1)
    l.append(x)
```

```python
print(l)
```

```python
plt.plot([i for i in range(20)],l)
```

```python
plt.hist(l)
```

```python
arr=[]
for i in l:
    x=math.sin(i)
    arr.append(x)
    print(x)
```

```python
arr
```

```python
plt.plot([i for i in range(20)],arr)
```
